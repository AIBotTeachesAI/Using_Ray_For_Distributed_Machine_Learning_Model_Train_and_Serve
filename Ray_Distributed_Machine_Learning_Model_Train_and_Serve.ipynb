### **What is Ray?**

**Ray** is an open-source framework designed to simplify the development of distributed applications. It provides a unified API for running tasks in a distributed fashion, allowing developers to scale their Python applications across multiple machines with ease. Ray's design makes it particularly well-suited for tasks that require parallel execution, such as machine learning, data processing, and reinforcement learning.

#### **Key Features of Ray:**
1. **Distributed Computing**:
   - Ray enables easy parallelization of tasks across multiple cores or machines. It abstracts away the complexity of distributed systems, allowing developers to write code as if it's running on a single machine while Ray handles the distribution.

2. **Fault Tolerance**:
   - Ray is built with fault tolerance in mind. It ensures that if a node in a distributed system fails, the tasks that were running on that node can be recovered and redistributed across the remaining nodes.

3. **Scalability**:
   - Ray scales seamlessly from a single laptop to a large cluster of machines in the cloud. This makes it ideal for applications that need to handle large-scale data processing or complex machine learning models.

4. **Ecosystem and Libraries**:
   - Ray comes with a rich ecosystem of libraries tailored for different use cases:
     - **Ray Tune** for hyperparameter tuning.
     - **Ray Serve** for scalable model serving.
     - **Ray Train** for distributed model training.
     - **Ray RLlib** for reinforcement learning.
     - **Ray Datasets** for distributed data processing.

5. **Ease of Use**:
   - Ray provides simple Python APIs that allow developers to scale their applications without needing to know the intricacies of distributed computing. This ease of use lowers the barrier to entry for scaling applications.

### **Ray Distributed Train**

**Ray Distributed Train** is a part of Ray's ecosystem designed to facilitate the distributed training of machine learning models. It allows developers to scale their model training across multiple GPUs and nodes with minimal code changes.

#### **How Ray Distributed Train Works:**
1. **Parallel Execution**:
   - Ray Distributed Train can distribute the training process across multiple GPUs or machines. This parallel execution speeds up the training process by leveraging the computational power of multiple devices.

2. **Resource Management**:
   - Ray efficiently manages the resources (like CPUs, GPUs, and memory) across the cluster, ensuring that the training process utilizes the available hardware optimally.

3. **Fault Tolerance**:
   - In case of node failures, Ray can recover the training process by redistributing the workload across the remaining nodes, minimizing disruptions.

4. **Framework Support**:
   - Ray Distributed Train is framework-agnostic, meaning it can work with popular machine learning frameworks like TensorFlow, PyTorch, and others. This flexibility allows developers to use their preferred tools while still benefiting from distributed training.

5. **Integration with Ray Tune**:
   - Ray Distributed Train integrates seamlessly with Ray Tune, allowing developers to perform hyperparameter tuning in a distributed manner, further optimizing their models.

### **Ray Serve**

**Ray Serve** is another component of the Ray ecosystem, designed for scalable and efficient model serving. It allows you to deploy machine learning models as web services that can handle high-throughput requests with low latency.

#### **Key Features of Ray Serve:**
1. **Scalable Model Serving**:
   - Ray Serve can scale horizontally, meaning it can distribute the load across multiple machines or instances, making it capable of handling a large number of requests simultaneously.

2. **Dynamic Batching**:
   - Ray Serve can dynamically batch incoming requests to optimize the utilization of resources, improving throughput and reducing latency.

3. **Model Composition**:
   - Ray Serve allows developers to compose multiple models or business logic into a single deployment pipeline, enabling complex inference workflows where multiple models may need to be invoked in sequence or parallel.

4. **Autoscaling**:
   - Ray Serve can automatically scale the number of replicas (instances) of your service up or down based on the incoming traffic, ensuring that the system remains responsive while also being cost-effective.

5. **Easy Integration**:
   - Ray Serve integrates smoothly with other Ray components, such as Ray Distributed Train, allowing you to easily transition from model training to deployment without major code changes.

### **Summary**

Ray provides a powerful, flexible, and easy-to-use framework for building distributed applications. With Ray Distributed Train, you can scale your machine learning model training across multiple GPUs and nodes, while Ray Serve allows you to deploy those models efficiently, handling high-throughput requests with ease. Together, these components make Ray an ideal choice for developing, training, and deploying machine learning models at scale.

```python
import os
import tempfile

import torch
from torch.nn import CrossEntropyLoss
from torch.optim import Adam
from torch.utils.data import DataLoader
from torchvision.models import resnet18
from torchvision.datasets import FashionMNIST
from torchvision.transforms import ToTensor, Normalize, Compose

import ray.train.torch

def train_func():
    # Model, Loss, Optimizer
    model = resnet18(num_classes=10)
    model.conv1 = torch.nn.Conv2d(
        1, 64, kernel_size=(7, 7), stride=(2, 2), padding=(3, 3), bias=False
    )
    # [1] Prepare model.
    model = ray.train.torch.prepare_model(model)
    # model.to("cuda")  # This is done by `prepare_model`
    criterion = CrossEntropyLoss()
    optimizer = Adam(model.parameters(), lr=0.001)

    # Data
    transform = Compose([ToTensor(), Normalize((0.5,), (0.5,))])
    data_dir = os.path.join(tempfile.gettempdir(), "data")
    train_data = FashionMNIST(root=data_dir, train=True, download=True, transform=transform)
    train_loader = DataLoader(train_data, batch_size=128, shuffle=True)
    # [2] Prepare dataloader.
    train_loader = ray.train.torch.prepare_data_loader(train_loader)

    # Training
    for epoch in range(3):
        if ray.train.get_context().get_world_size() > 1:
            train_loader.sampler.set_epoch(epoch)

        for images, labels in train_loader:
            # This is done by `prepare_data_loader`!
            # images, labels = images.to("cuda"), labels.to("cuda")
            outputs = model(images)
            loss = criterion(outputs, labels)
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

        # [3] Report metrics and checkpoint.
        metrics = {"loss": loss.item(), "epoch": epoch}
        with tempfile.TemporaryDirectory() as temp_checkpoint_dir:
            # Handle the case where the model is not wrapped in DistributedDataParallel
            if hasattr(model, 'module'):
                torch.save(
                    model.module.state_dict(),
                    os.path.join(temp_checkpoint_dir, "model.pt")
                )
            else:
                torch.save(
                    model.state_dict(),
                    os.path.join(temp_checkpoint_dir, "model.pt")
                )
            ray.train.report(
                metrics,
                checkpoint=ray.train.Checkpoint.from_directory(temp_checkpoint_dir),
            )
        if ray.train.get_context().get_world_rank() == 0:
            print(metrics)

# [4] Configure scaling and resource requirements.
#scaling_config = ray.train.ScalingConfig(num_workers=2, use_gpu=True)
#above did not work in colab
scaling_config = ray.train.ScalingConfig(num_workers=1, use_gpu=True)


# [5] Launch distributed training job.
trainer = ray.train.torch.TorchTrainer(
    train_func,
    scaling_config=scaling_config,
    # [5a] If running in a multi-node cluster, this is where you
    # should configure the run's persistent storage that is accessible
    # across all worker nodes.
    # run_config=ray.train.RunConfig
        # run_config=ray.train.RunConfig(storage_path="s3://..."),
)
result = trainer.fit()
```

```python

# [6] Load the trained model.
with result.checkpoint.as_directory() as checkpoint_dir:
    model_state_dict = torch.load(os.path.join(checkpoint_dir, "model.pt"))
    model = resnet18(num_classes=10)
    model.conv1 = torch.nn.Conv2d(
        1, 64, kernel_size=(7, 7), stride=(2, 2), padding=(3, 3), bias=False
    )
    model.load_state_dict(model_state_dict)
```

```python
import requests
from starlette.requests import Request
from typing import Dict
import threading
import ray
from ray import serve
import time

# Initialize Ray
ray.init(ignore_reinit_error=True)

# 1: Define a Ray Serve application.
@serve.deployment
class MyModelDeployment:
    def __init__(self, msg: str):
        self._msg = msg

    async def __call__(self, request: Request) -> Dict:
        return {"result": self._msg}

app = MyModelDeployment.bind(msg="Hello world!")

# 2: Deploy the application locally.
serve.run(app, route_prefix="/")

# Function to start the Ray Serve application
def run_serve():
    serve.run(app, route_prefix="/")
    while True:
        time.sleep(10)

# Run the Ray Serve application in a background thread
threading.Thread(target=run_serve, daemon=True).start()

# Give it some time to start up
time.sleep(5)


```

```python
import os
import tempfile

import torch
from torch.nn import CrossEntropyLoss
from torch.optim import Adam
from torch.utils.data import DataLoader
from torchvision.models import resnet18
from torchvision.datasets import CIFAR10
from torchvision.transforms import ToTensor, Normalize, Compose

import ray.train.torch

def train_func():
    # Model, Loss, Optimizer
    model = resnet18(num_classes=10)
    # Since CIFAR-10 images have 3 channels, no need to modify the first convolutional layer.
    model = ray.train.torch.prepare_model(model)
    criterion = CrossEntropyLoss()
    optimizer = Adam(model.parameters(), lr=0.001)

    # Data
    transform = Compose([
        ToTensor(),
        Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
    ])
    data_dir = os.path.join(tempfile.gettempdir(), "data")
    train_data = CIFAR10(root=data_dir, train=True, download=True, transform=transform)
    train_loader = DataLoader(train_data, batch_size=128, shuffle=True)
    train_loader = ray.train.torch.prepare_data_loader(train_loader)

    # Training
    for epoch in range(30):
        if ray.train.get_context().get_world_size() > 1:
            train_loader.sampler.set_epoch(epoch)

        for images, labels in train_loader:
            outputs = model(images)
            loss = criterion(outputs, labels)
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

        # Report metrics and checkpoint.
        metrics = {"loss": loss.item(), "epoch": epoch}
        with tempfile.TemporaryDirectory() as temp_checkpoint_dir:
            if hasattr(model, 'module'):
                torch.save(
                    model.module.state_dict(),
                    os.path.join(temp_checkpoint_dir, "model.pt")
                )
            else:
                torch.save(
                    model.state_dict(),
                    os.path.join(temp_checkpoint_dir, "model.pt")
                )
            ray.train.report(
                metrics,
                checkpoint=ray.train.Checkpoint.from_directory(temp_checkpoint_dir),
            )
        if ray.train.get_context().get_world_rank() == 0:
            print(metrics)

# Configure scaling and resource requirements.
scaling_config = ray.train.ScalingConfig(num_workers=1, use_gpu=True)

# Launch distributed training job.
trainer = ray.train.torch.TorchTrainer(
    train_func,
    scaling_config=scaling_config,
)
result = trainer.fit()
```

```python
import torch
from torchvision.models import resnet18
from torchvision.transforms import Compose, Normalize, ToTensor
from starlette.requests import Request
from typing import Dict
from ray import serve
import ray
import os

# Initialize Ray
ray.init(ignore_reinit_error=True)

# 1: Define a Ray Serve application.
@serve.deployment
class MyModelDeployment:
    def __init__(self, model_checkpoint_dir: str):
        # Load the trained model
        self.model = resnet18(num_classes=10)
        checkpoint_path = os.path.join(model_checkpoint_dir, "checkpoint_000029", "model.pt")

        # Ensure CUDA is available
        device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

        # Load the model with map_location to cuda
        self.model.load_state_dict(torch.load(checkpoint_path, map_location=device))
        self.model.to(device)
        self.model.eval()  # Set model to evaluation mode
        self.transform = Compose([
            Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
        ])
        self.device = device

    async def __call__(self, request: Request) -> Dict:
        # Get the input data from the request
        data = await request.json()
        image_tensor = torch.tensor(data['image']).permute(2, 0, 1).unsqueeze(0).float().to(self.device)
        image_tensor = self.transform(image_tensor)

        # Make a prediction
        with torch.no_grad():
            prediction = self.model(image_tensor)
            predicted_class = torch.argmax(prediction, dim=1).item()

        return {"predicted_class": predicted_class}

# Specify the directory where your model checkpoint is stored
model_checkpoint_dir = "/root/ray_results/TorchTrainer_2024-08-19_17-33-30/TorchTrainer_26727_00000_0_2024-08-19_17-33-30/"

app = MyModelDeployment.bind(model_checkpoint_dir=model_checkpoint_dir)

# 2: Deploy the application locally.
serve.run(app, route_prefix="/")
```
